```python
def one_word_list(sent_list):
    one_list=[]
    for sentence in sent_list:
        one_list.extend(sentence)
    return(one_list)

review_df['comment_n']=review_df['comment_n'].apply(lambda x: one_word_list(eval(x)))
review_df['comment_n']
```

```python
review_df['comment_n_str']=review_df['comment_n'].apply(lambda x: " ".join(x))
review_df['comment_n_str']
```

```python
# CountVectorizer 사용
from sklearn.feature_extraction.text import CountVectorizer

vector = CountVectorizer()
count_dtm=vector.fit_transform(review_df['comment_n_str']).toarray() 
print(count_dtm.shape)
print(vector.vocabulary_)
```

```python
# TfidfVectorizer 사용
from sklearn.feature_extraction.text import TfidfVectorizer

tfidfv = TfidfVectorizer().fit(review_df['comment_n_str'])
tfidv_dtm=tfidfv.transform(review_df['comment_n_str']).toarray()
print(tfidv_dtm.shape)
print(tfidfv.vocabulary_)
```
