```python
# Îã®Ïñ¥ ÏÇ¨Ï†ÑÍ≥º Bag of Words  Î≥ÄÌôò

from gensim import corpora
dictionary = corpora.Dictionary(review_df['comment_n'])
print(dictionary)
```

```python
corpus = [dictionary.doc2bow(text) for text in review_df['comment_n']]

print(corpus[0])

for i in range(5):
    print(dictionary[i])
```

```python
import gensim
NUM_TOPICS = 10 #10Í∞úÏùò ÌÜ†ÌîΩ, k=10
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)

topics = ldamodel.print_topics(num_words=4)
print(topics)
```

```python
# sklearnÏóêÏÑú LDA

from sklearn.feature_extraction.text import CountVectorizer

# the vectorizer object will be used to transform text to vector form
vectorizer = CountVectorizer(max_df=0.9, min_df=25)

# apply transformation
review_df['comment_n_str']=[' '.join(token) for token in review_df['comment_n']]
tf = vectorizer.fit_transform(review_df['comment_n_str']).toarray()

# tf_feature_names tells us what word each column in the matric represents
tf_feature_names = vectorizer.get_feature_names()
```

```python
from sklearn.decomposition import LatentDirichletAllocation

number_of_topics = 10

model = LatentDirichletAllocation(n_components=number_of_topics, random_state=0)

model.fit(tf)
```

```python
# ÌÜ†ÌîΩ Î≥¥Í∏∞
def display_topics(model, feature_names, no_top_words):
    topic_dict = {}
    for topic_idx, topic in enumerate(model.components_):
        topic_dict["Topic %d words" % (topic_idx)]= ['{}'.format(feature_names[i])
                        for i in topic.argsort()[:-no_top_words - 1:-1]]
        topic_dict["Topic %d weights" % (topic_idx)]= ['{:.1f}'.format(topic[i])
                        for i in topic.argsort()[:-no_top_words - 1:-1]]
    return pd.DataFrame(topic_dict)
```

```python
no_top_words = 10
topic_dictionary=display_topics(model, tf_feature_names, no_top_words)
```
